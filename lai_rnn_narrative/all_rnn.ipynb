```python
def get_tracklists(df):
    tracklist = []
    album = ""
    for i, track in df.iterrows():
        if album == "":
            album = track['release_x']
        if track['release_x'] != album:
            yield album, tracklist
            album = track['release_x']
            tracklist = []
        tracklist += [track]
    yield album, tracklist
```

```python
from math import isnan
default_vector = [0 for f in range(amounts['all'])]
"""edit default vector here"""

def to_vector(track):
    strs = [track[f] for f in features['all']]
    fs = [float(bool(s)) if s in ['True','False']  
                else float(s) for s in strs]
    return [default_vector[i] if isnan(fs[i]) else fs[i] for i in range(amounts['all'])]
```

```python
import numpy as np
metric_scaler = {f : 1.0 for f in features['all']}
"""set scalers here"""
metric_scaler['loudness'] = 0.1
metric_scaler['duration'] = 0.001
metric_scaler['tempo'] = 0.1
metric_scaler['time_signature'] = 0.5
metric_scaler['key'] = 0.3
metric_scaler = np.array([metric_scaler[f] for f in features['all']])
```

```python
def metric(vector1, vector2):
    v1, v2 = np.array(vector1), np.array(vector2)
    d = (v1 - v2)
    d = d * metric_scaler
    m = d * d
    return np.sum(m)
```

```python
def to_matrix(tracklist):
    return [to_vector(track) for track in tracklist]

def seq_matrices(df):
    for album, tracklist in get_tracklists(df):
        yield to_matrix(tracklist)

def group_matrices_by_length(df):
    groups = {}
    for m in seq_matrices(df):
        if len(m) in groups:
            groups[len(m)] += [m]
        else:
            groups[len(m)] = [m]
    return groups
```

```python
groups = group_matrices_by_length(dfs['all'])
```

```python

from keras.models import Sequential
from keras.layers import LSTM, Dropout, Dense

model = Sequential()
# batch, time_steps, features
# input_shape = (time_steps, features)
# time_steps = variable = -1 or None
model.add(LSTM(200, input_shape=(None, amounts['all']), return_sequences=True))
model.add(Dropout(0.5))
model.add(LSTM(200, return_sequences=True))
model.add(Dense(amounts['all']))
# no activation function, because features are not normalized
print(model.summary())

model.compile(optimizer="adam", loss="MSE")
```

```python
def train(model,epochs):
    for epoch in range(epochs):
        print("epoch",epoch)
        for l, matrices in groups.items():
            print("tracklists of length", l, "amount", len(matrices))
            if len(matrices) < 1 or l <= 1: continue
            batch = np.array(matrices)
            x = batch[:,:-1]
            y = batch[:,1:]
            model.fit(x,y,epochs = 1, verbose = 0)
```

```python
%%time

# train RNN
train(model,1)
```
